Queue

```python
# Simplified implementation (relying on built-in data structures)

class Queue:
    def __init__(self):
        self.items = []

    def enqueu(self, value):
        self.items.insert(0, value)

    def dequeu(self):
        return self.items.pop()

    # nice to hae methods
    def size(self):
        return len(self.items)

    def peek(self):
        return self.items[len(self.items)-1]

    def is_empty(self):
        return self.items == []
```
